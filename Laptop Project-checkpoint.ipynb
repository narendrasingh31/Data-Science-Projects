```python
def header():
    print('-'*80)
    print('Welcome to Ethans shop'.center(80,' '))
    print('-'*80)

def laptop():
    brand_price,brand = get_brand()
    ram_price,ram = get_ram()
    drive_price = get_drive()
    processor_price = get_processor_type()
    screen_size_price = get_screen_size()
    c_configuration()
    total_laptop_price = brand_price + ram_price + drive_price + processor_price + screen_size_price
    print('-'*80)
    print(f'Total price of your {brand} Laptop is :{total_laptop_price}'.rjust(80,' '))
    
    print('-'*80)
    print('.....Thank You for visiting us.....'.center(80,' '))


def c_configuration():
    header()
    print('You have purchased laptop with following configuration :-\n')
    print(f"{brand} {ram} RAM , {size} {drive}\n{processor}, {screen_size} screen")
    
def get_brand():
    header()
    print('Please choose Laptop Brand')
    print('1 .DELL\n2 .HP\n3 .ASUS\n4 .LENOVA')
    global brand
    brand = ''
    
    while True:
        option = input('Enter your choice :')
        if not option.isdigit():
            print('ERROR!! Please enter only digits'.center(80,' '))
        else:
            break
    price = 0
    if option == '1':
        brand = 'DELL'
        price = 22000
    elif option == '2':
        brand = 'HP'
        price = 20000
    elif option == '3':
        brand = 'ASUS'
        price = 17000
    elif option == '4':
        brand = 'LENOVA'
        price = 18000
    else:
        print('ERROR!! please choose right option'.center(80,' '))
        get_brand()
    return price,brand


def get_ram():
    header()
    print('Please choose RAM')
    print('1 .4GB\n2 .8GB\n3 .16GB')
    global ram
    ram = ''
    while True:
        option = input('Enter your choice :')
        if not option.isdigit():
            print('ERROR!! Please enter only digits'.center(80,' '))
        else:
            break
    price = 0
    if option == '1':
        ram = '4GB'
        price = 5000
    elif option == '2':
        ram = '8GB'
        price = 15000
    elif option == '3':
        ram = '16GB'
        price = 22000
    else:
        print('ERROR!! please choose right option'.center(80,' '))
        get_ram()
    return price,ram

def get_drive():
    header()
    print('Please choose storage drive')
    print('1 .SSD\n2 .HDD\n3 .HYBRID')
    global drive,size
    price =''
    while True:
        option = input('Enter your choice :')
        if not option.isdigit():
            print('ERROR!! Please enter only digits'.center(80,' '))
        else:
            break
    if option == '1':
        drive = 'SSD'
        price,size = get_ssd()
    elif option == '2':
        drive = 'HDD'
        price,size = get_hdd()
    elif option == '3':
        drive = 'HYBRID'
        price,size = get_hybrid()
    else:
        print('ERROR!! please choose right option'.center(80,' '))
        get_drive()
    return price


def get_ssd():
    header()
    print('Please choose SSD storage')
    print('1 .256GB\n2 .512GB\n3 .1TB')
    while True:
        option = input('Enter your choice :')
        if not option.isdigit():
            print('ERROR!! Please enter only digits'.center(80,' '))
        else:
            break
    price = 0
    if option == '1':
        ssd = '256GB'
        price = 1700
    elif option == '2':
        ssd = '512GB'
        price = 3200
    elif option == '3':
        ssd = '1TB'
        price = 5200
    else:
        print('ERROR!! please choose right option'.center(80,' '))
        get_ssd()
    return price,ssd

def get_hdd():
    header()
    print('Please choose HDD storage')
    print('1 .512GB\n2 .1TB')
    while True:
        option = input('Enter your choice :')
        if not option.isdigit():
            print('ERROR!! Please enter only digits'.center(80,' '))
        else:
            break
    price = 0
    if option == '1':
        hdd = '512GB'
        price = 850
    elif option == '2':
        hdd = '1TB'
        price = 1250
    else:
        print('ERROR!! please choose right option'.center(80,' '))
        get_hdd()
    return price,hdd

def get_hybrid():
    header()
    print('Please choose HYBRID storage')
    print('1 .256GB SSD + 1TB HDD\n2 .512GB SSD + 1TB HDD')
    while True:
        option = input('Enter your choice :')
        if not option.isdigit():
            print('ERROR!! Please enter only digits'.center(80,' '))
        else:
            break
    price = 0
    if option == '1':
        hybrid = '256GB SSD + 1TB HDD'
        price = 2850
    elif option == '2':
        hybrid = '512GB SSD + 1TB HDD'
        price = 3450
    else:
        print('ERROR!! please choose right option'.center(80,' '))
        get_hybrid()
    return price,hybrid

def get_processor_type():
    header()
    print('Please choose Processor Type')
    print('1 .Ryzen\n2 .intel core')
    global processor
    price = ''
    while True:
        option = input('Enter your choice :')
        if not option.isdigit():
            print('ERROR!! Please enter only digits'.center(80,' '))
        else:
            break
    if option == '1':
        processor_type = 'Ryzen'
        price,processor = get_ryzen()
    elif option == '2':
        processor_type = 'intel core'
        price,processor = get_intel_core()
    else:
        print('ERROR!! please choose right option'.center(80,' '))
        get_processor_type()
    return price

def get_ryzen():
    header()
    print('Please choose Ryzen type processor')
    print('1 .Ryzen-3\n2 .Ryzen-5\n3 .Ryzen-7')
    while True:
        option = input('Enter your choice :')
        if not option.isdigit():
            print('ERROR!! Please enter only digits'.center(80,' '))
        else:
            break
    price = 0
    ryzen = ''
    if option == '1':
        ryzen = 'Ryzen-3'
        price = 1000
    elif option == '2':
        ryzen = 'Ryzen-5'
        price = 3800
    elif option == '3':
        ryzen = 'Ryzen-7'
        price = 6100
    else:
        print('ERROR!! please choose right option'.center(80,' '))
        get_ryzen()
    return price,ryzen

def get_intel_core():
    header()
    print('Please choose Intel Core type processor')
    print('1 .i-3\n2 .i-5\n3 .i-7')
    while True:
        option = input('Enter your choice :')
        if not option.isdigit():
            print('ERROR!! Please enter only digits'.center(80,' '))
        else:
            break
    intel_core = ''
    price = 0
    if option == '1':
        intel_core = 'i-3'
        price = 1350
    elif option == '2':
        intel_core = 'i-5'
        price = 4500
    elif option == '3':
        intel_core = 'i-7'
        price = 7000
    else:
        print('ERROR!! please choose right option'.center(80,' '))
        get_intel_core()
    return price,intel_core

def get_screen_size():
    header()
    print('Please choose Screen size')
    global screen_size
    print('1 .34.04 cm (13.4 inch)\n2 .35.56 cm (14 inch)\n3 .39.62 cm (15.6 inch)')
    while True:
        option = input('Enter your choice :')
        if not option.isdigit():
            print('ERROR!! Please enter only digits'.center(80,' '))
        else:
            break
    price = 0
    if option == '1':
        screen_size = '34.04 cm (13.4 inch)'
        price = 1800
    elif option == '2':
        screen_size = '35.56 cm (14 inch)'
        price = 2950
    elif option == '3':
        screen_size = '39.62 cm (15.6 inch)'
        price = 4200
    else:
        print('ERROR!! please choose right option'.center(80,' '))
        get_screen_size()
    return price
```
